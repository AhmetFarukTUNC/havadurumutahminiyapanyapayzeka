# Train a Classification Model

```python


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)


# Create and train the model
model = RandomForestClassifier(random_state=42)



model.fit(X_train, y_train)
```

# Make Predictions

```python
# Predict the test set results

y_pred = model.predict(X_test)
```

# Printing y_pred

```python
y_pred
```

# Evaluate the Model

```python
# Evaluate the model
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

# Calculate ROC Metrics:

# You will add description.

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
import matplotlib.pyplot as plt
import numpy as np

# Ensure that the classes are sorted and labeled correctly
classes = np.unique(y_test)
y_test_bin = label_binarize(y_test, classes=classes)

# Fit the model using One-vs-Rest strategy
ovr = OneVsRestClassifier(model)
ovr.fit(X_train, y_train)
y_probs = ovr.predict_proba(X_test)

# Initialize the plot
plt.figure(figsize=(8, 6))

# Function to plot ROC curves for each class
def plot_roc_curve(y_true, y_prob, class_name):
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{class_name} (AUC = {roc_auc:.2f})')

# Loop through each class and plot its ROC curve
for i, class_name in enumerate(classes):
    plot_roc_curve(y_test_bin[:, i], y_probs[:, i], f'Class {class_name}')

# Final plot setup
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for reference
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Multiclass Classification')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
